```python
#queremos filtrar solo los regustros de DartVader

d_star_wars['personaje']=='Darth Vader'

#usamos el == para una coparación
```

```python
#le cambiamos la altura a Darth Vader
v_flag=d_star_wars['personaje']=='Darth Vader'
```

```python
#Como cambiar un valor basado en una condición

d_star_wars.loc[v_flag,:] #filtra solo el regustro de Darth Vader
```

```python
d_star_wars.loc[v_flag,'altura']=185  #en todos los registros dond ela variable v_flag sea true se va a remmplazar la altura por 185
d_star_wars
```

```python
#también sirve para crear una variable nueva

d_star_wars['altura2']=185
d_star_wars
```

```python
#ahora hacemos todo en la misma línea per para Luke Skywaker
d_star_wars.loc[d_star_wars['personaje']=='Luke Skywaker','altura']=200
d_star_wars
```
